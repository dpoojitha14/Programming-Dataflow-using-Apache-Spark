Plotting the performance of Apache Spark infrastructure.

The time taken for bigrams for n files is as follows:
    
    n    Time Taken (in seconds)
    1    06.863662
    10   09.135117
    30   17.026222
    50   24.848370
    100  44.185310
    120  55.287645

The time taken for trigrams for n files is as follows:
   
    n    Time Taken (in seconds)
    1    05.413452
    10   06.613591
    30   08.818708
    50   11.091574
    100  18.067519
    120  21.369038

Plotting the time taken vs number of files:

```r
files = c(1,10,30,50,100,120)
bigrams = c(06.863662,09.135117,17.026222,24.848370,44.185310,55.287645)
trigrams = c(05.413452,06.613591,08.818708,11.091574,18.067519,21.369038)
```

```r
plot(files, bigrams, type="l",col= "red",xlab = "No of files", ylab = "Time Taken",main = "Performance of Apache Spark infrastructure \n (Red:Bigrams, Blue:Trigrams)")
lines(files,trigrams,col = "blue")
```

Comments:

It can be observed that the time taken increases as the number of files increase.

The time taken for trigrams is less than the time taken for bigrams when the same number of files are considered.
